###**Ініціалізація ваг**

Ініціалізація ваг у нейронних мережах є одним із критичних етапів у процесі
розробки та навчання моделей. Від того, як ваги будуть задані на старті,
безпосередньо залежить, як швидко модель збіжиться до оптимального рішення,
або навіть чи збіжиться взагалі.


Декілька ключових моментів, що підкреслюють важливість
правильної ініціалізації:

- ***Мінімізація ризику затухання або вибуху градієнтів:*** При дуже малих
початкових вагах градієнти можуть зменшуватися експоненційно під час
поширення, призводячи до затухання градієнтів. З іншого боку, завеликі
ваги можуть призводити до вибуху градієнтів, коли значення градієнтів
стають занадто великими. В обох випадках навчання може стати
неефективним або неможливим.
- ***Швидкість збіжності:*** Правильно задані ваги можуть сприяти швидшому
навчанню моделі, тому що оптимальні ваги знаходяться швидше. Це
економить час та ресурси при тренуванні складних моделей.
- ***Уникнення застрягання в локальних мінімумах:*** Хоча глобальний мінімум
в функції втрат є ідеальною метою, існують численні локальні мінімуми та
сідлові точки. Правильна ініціалізація може допомогти оптимізатору
уникати застрягання в неоптимальних рішеннях.


###***Нормалізація даних***


Нормалізація даних – це процес масштабування вхідних змінних до певного
діапазону, зазвичай в діапазоні від 0 до 1 або від -1 до 1. Цей процес може бути
дуже корисним при навчанні нейронних мереж, і ось чому:

- ***Швидкість збіжності:*** Нормалізація може сприяти швидшому навчанню
моделі. Коли всі вхідні функції мають масштаб, що приблизно однаковий,
градієнтний спуск збіжиться швидше, адже усі ваги будуть оновлюватися
приблизно однаковими кроками.
- ***Запобігання затуханню чи вибуху градієнтів:*** Якщо одна змінна має дуже
великий масштаб порівняно з іншими, це може призвести до ситуації, коли
ваги для цієї змінної будуть оновлюватися набагато швидше, ніж для
інших. Це може призвести до нестабільного навчання, вибуху або
затухання градієнтів.
- ***Покращення загальної продуктивності моделі:*** Моделі часто досягають
кращої продуктивності, коли їх вхідні дані нормалізовано. Наприклад,
мережі з активаційними функціями, такими як сигмоїда або гіперболічний
тангенс, працюють найкраще, коли вхідні дані центровані навколо нуля.
- ***Поєднання різних масштабів змінних:*** У реальних наборах даних різні
змінні можуть мати дуже різні масштаби. Наприклад, вік може бути
представлений двозначним числом, а дохід – шестизначним. Нормалізуючи
обидва цих атрибути, ми даємо їм рівні шанси впливати на навчання
моделі.
- ***Більш інтерпретовані ваги:*** Коли вхідні дані нормалізовано, ваги, які
модель призначає різним змінним, можуть бути інтерпретовані як міра
важливості цих змінних.

###***Регуляризації та Dropout у запобіганні перенавчанню !!!***

Перенавчання (або overfitting) є однією з найбільших проблем при навчанні
нейронних мереж. Це стається, коли модель занадто добре навчається на
тренувальному наборі даних і стає занадто специфічною для нього, втрачаючи
при цьому здатність узагальнювати результати на нових даних. Регуляризація та
Dropout — це два підходи, які допомагають запобігти перенавчанню.


###***Регуляризація.***

Регуляризація вводить штраф для великих ваг моделі, роблячи її менш складною.
Існує два основних типи регуляризації: L1 та L2. L1 регуляризація додає штраф,
пропорційний абсолютному значенню ваг, тоді як L2 регуляризація додає штраф,
пропорційний квадрату значення ваг.
Регуляризація змушує модель мати менші ваги, що робить її менш чутливою до
окремих вхідних прикладів, сприяючи кращій генералізації.


###***Dropout.***

Dropout – це техніка, при якій випадково "відключаються" (або встановлюються
в нуль) певні нейрони під час навчання. Це означає, що під час проходження
вперед кожного міні-батча вхід деяких нейронів встановлюється на 0. Це робить
модель менш залежною від конкретних нейронів, сприяючи розподілу
відповідальності по всій мережі.

Така "розрідженість" мережі змушує її вчитися більш стійким (robust) шляхом,
оскільки вона не може покладатися на будь-який конкретний нейрон.

Обидва цих підходи працюють за принципом введення деякої форми шуму або
обмеження в процес навчання, що сприяє кращій генералізації моделі та
зменшенню перенавчання.

Використовуючи регуляризацію та Dropout у комбінації, можна значно
покращити стійкість моделі до перенавчання і підвищити її продуктивність на
тестових наборах даних.

###Крок 2: Ініціалізація ваг

Давайте спробуємо два популярні методи:

- **Xavier (Glorot)** — добре працює з функціями активації, як-от tanh.

- **He** — рекомендований для функцій активації типу ReLU.

Модель буде побудована в середовищі Keras.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.initializers import GlorotNormal, HeNormal
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error

# Функція для створення моделі
def build_model(initializer, input_dim):
    model = Sequential()
    model.add(Dense(64, activation='relu', kernel_initializer=initializer, input_dim=input_dim))
    model.add(Dense(32, activation='relu', kernel_initializer=initializer))
    model.add(Dense(1))  # Лінійний вихід для регресії
    model.compile(optimizer=Adam(learning_rate=0.01), loss='mse', metrics=['mae'])
    return model

# Створення моделей із різними ініціалізаторами
model_xavier = build_model(GlorotNormal(), X_train.shape[1])
model_he = build_model(HeNormal(), X_train.shape[1])

# Навчання моделей
history_xavier = model_xavier.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=0)
history_he = model_he.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=0)

# Оцінка на тестових даних
y_pred_xavier = model_xavier.predict(X_test)
y_pred_he = model_he.predict(X_test)

mse_xavier = mean_squared_error(y_test, y_pred_xavier)
mse_he = mean_squared_error(y_test, y_pred_he)

print("MSE (Xavier):", mse_xavier)
print("MSE (He):", mse_he)
```

```python
# Функція для візуалізації втрат
def plot_loss(history, title):
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(title)
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

# Візуалізація
plot_loss(history_xavier, "Xavier Initialization")
plot_loss(history_he, "He Initialization")
```

####***Опис результатів***

**Параметри навчання:**

Було створено дві моделі нейронних мереж для розв’язання задачі регресії.

Основна різниця між моделями — метод ініціалізації ваг:

- *Xavier (Glorot Normal):* спрямований на збалансування дисперсії вхідних і вихідних даних для кожного шару.

- *He Normal:* спеціалізований для активаційних функцій типу ReLU, враховуючи властивості градієнтів.

**Гіперпараметри:**

Кількість шарів: 3 (2 приховані шари з 64 та 32 нейронами відповідно, один вихідний шар).

Активаційна функція: ReLU для прихованих шарів, лінійна для вихідного.

Оптимізатор: Adam із навчальною швидкістю
0.01.

Кількість епох: 50.

Batch size: 32.

Відсоток даних для валідації: 20%.

**Результати:**

MSE (середньоквадратична помилка) на тестових даних:

- Xavier: 2790.22.
- He: 2793.75.

Обидва методи ініціалізації показали близькі результати, що свідчить про схожу ефективність для даного набору даних.


Також наведені графіки навчання (Train Loss) та валідації (Validation Loss) показують, як змінювалася функція втрат протягом епох.

*Графік 1: Xavier Initialization*

Плавна збіжність.
Модель досягає стабільного значення після приблизно 30 епох.

*Графік 2: He Initialization*

Маємо хожий характер збіжності, хоча в деяких епохах втрати валідації трохи більші.

Крок: 3

```python
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import Dropout

# Функція для створення моделі з L2 регуляризацією та Dropout
def build_model_with_regularization(initializer, input_dim, dropout_rate=0.5, l2_penalty=0.01):
    model = Sequential()
    model.add(Dense(64, activation='relu', kernel_initializer=initializer, input_dim=input_dim, kernel_regularizer=l2(l2_penalty)))
    model.add(Dropout(dropout_rate))  # Додаємо Dropout після першого шару
    model.add(Dense(32, activation='relu', kernel_initializer=initializer, kernel_regularizer=l2(l2_penalty)))
    model.add(Dropout(dropout_rate))  # Додаємо Dropout після другого шару
    model.add(Dense(1))  # Лінійний вихід для регресії
    model.compile(optimizer=Adam(learning_rate=0.01), loss='mse', metrics=['mae'])
    return model

# Створення моделей з регуляризацією L2 і Dropout
model_xavier_reg = build_model_with_regularization(GlorotNormal(), X_train.shape[1])
model_he_reg = build_model_with_regularization(HeNormal(), X_train.shape[1])

# Навчання моделей
history_xavier_reg = model_xavier_reg.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=0)
history_he_reg = model_he_reg.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=0)

# Оцінка на тестових даних
y_pred_xavier_reg = model_xavier_reg.predict(X_test)
y_pred_he_reg = model_he_reg.predict(X_test)

mse_xavier_reg = mean_squared_error(y_test, y_pred_xavier_reg)
mse_he_reg = mean_squared_error(y_test, y_pred_he_reg)

print("MSE (Xavier з регуляризацією):", mse_xavier_reg)
print("MSE (He з регуляризацією):", mse_he_reg)
```

Аналіз результатів:
Xavier з регуляризацією показує кращий результат (менший MSE) порівняно з He з регуляризацією.
Можливо, He з регуляризацією не настільки ефективно працює для цього набору даних. Це може бути пов'язано з характеристиками функції активації ReLU або іншими факторами, які потребують додаткового налаштування.

Графіки втрат для кожної моделі:
Побудуємо графіки для Xavier з регуляризацією та He з регуляризацією. Це дозволить побачити, як змінилась функція втрат на тренувальних і валідаційних даних.

```python
# Функція для візуалізації навчальних втрат
def plot_loss_with_regularization(history, title):
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(title)
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

# Візуалізація
plot_loss_with_regularization(history_xavier_reg, "Xavier Initialization with Regularization")
plot_loss_with_regularization(history_he_reg, "He Initialization with Regularization")
```

Моделі з регуляризацією можуть працювати трохи гірше за ті, що без неї, оскільки вони вводять додаткові штрафи, які можуть не бути оптимальними для цього конкретного набору даних.
